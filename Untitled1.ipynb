```python
import pandas as pd
import geopandas
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_props = pd.read_csv('train.csv')
```

```python
df_props.info()
```

```python
# Optimizo el tipo de dato para aquellas columans que admitan valores booleanos
df_props['gimnasio'] = df_props['gimnasio'].astype('bool')
df_props['usosmultiples'] = df_props['usosmultiples'].astype('bool')
df_props['piscina'] = df_props['piscina'].astype('bool')
df_props['escuelascercanas'] = df_props['escuelascercanas'].astype('bool')
df_props['centroscomercialescercanos'] = df_props['centroscomercialescercanos'].astype('bool')
df_props.info()
```

```python
# Me quedo con las filas que tienen coordenadas validas
df_coordenadas = df_props[df_props['lat'].notnull() & df_props['lng'].notnull()]
df_coordenadas.info()
```
